For goal of predicting and reducing customer churn, here are the best chart types for each stage of analysis:
1. Customer Churn Overview
ðŸ“Š Key Goal: Identify the proportion of customers churning and their characteristics.

Pie Chart / Bar Chart: Proportion of churned vs. existing customers (Attrition_Flag).
Histogram / Box Plot: Compare Customer_Age, Months_on_book, and Total_Relationship_Count for churned vs. retained customers.
Stacked Bar Chart: Breakdown of churn rate by Gender, Income_Category, and Education_Level.
2. Customer Behavior & Churn Drivers
ðŸ“Š Key Goal: Identify trends and risk factors linked to churn.

Scatter Plot:

Total_Trans_Amt vs. Total_Trans_Ct to see if lower transaction frequency correlates with churn.
Credit_Limit vs. Avg_Utilization_Ratio to check if high utilization is linked to churn.
Box Plot:

Avg_Utilization_Ratio, Credit_Limit, and Months_Inactive_12_mon comparison between churned vs. retained customers.
Heatmap (Correlation Matrix): Identify relationships between features such as Months_Inactive_12_mon, Total_Trans_Ct, and Total_Amt_Chng_Q4_Q1 with churn.

Line Chart:

Total_Amt_Chng_Q4_Q1 vs. Total_Ct_Chng_Q4_Q1 to track spending and transaction changes leading to churn.
3. Predicting At-Risk Customers
ðŸ“Š Key Goal: Develop risk scores to proactively identify churn-prone customers.

Decision Tree / Sankey Diagram: Show top decision paths leading to churn (e.g., "If Months_Inactive_12_mon > X & Total_Trans_Ct < Y â†’ High Churn Risk").

Scatter Plot with Clustering: Cluster customers based on Credit_Limit, Total_Trans_Amt, Total_Ct_Chng_Q4_Q1, and Avg_Utilization_Ratio to separate churn-risk groups.

Cohort Analysis (Heatmap): Track customer churn over time based on their onboarding month (Months_on_book).


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("whitegrid")

# Pie Chart: Proportion of Churned vs. Existing Customers
plt.figure(figsize=(6, 6))
df_bankchurnersunclean["Attrition_Flag"].value_counts().plot.pie(autopct="%1.1f%%", startangle=90, colors=["lightblue", "salmon"])
plt.title("Customer Churn Distribution")
plt.ylabel("")
plt.show()

# Bar Chart: Churn Rate by Gender
plt.figure(figsize=(6, 4))
sns.countplot(data=df_bankchurnersunclean, x="Gender", hue="Attrition_Flag", palette=["lightblue", "salmon"])
plt.title("Churn Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Attrition Flag")
plt.show()

# Histogram: Age Distribution by Churn Status
plt.figure(figsize=(8, 5))
sns.histplot(data=df_bankchurnersunclean, x="Customer_Age", hue="Attrition_Flag", kde=True, bins=20, palette=["lightblue", "salmon"])
plt.title("Customer Age Distribution by Churn Status")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Box Plot: Months on Book for Churned vs. Retained Customers
plt.figure(figsize=(6, 5))
sns.boxplot(data=df_bankchurnersunclean, x="Attrition_Flag", y="Months_on_book", palette=["lightblue", "salmon"])
plt.title("Months on Book vs. Churn Status")
plt.xlabel("Churn Status")
plt.ylabel("Months on Book")
plt.show()
```

```python
# Scatter Plot: Total Transactions Amount vs. Total Transactions Count
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_bankchurnersunclean, x="Total_Trans_Ct", y="Total_Trans_Amt", hue="Attrition_Flag", alpha=0.7, palette=["lightblue", "salmon"])
plt.title("Total Transactions Count vs. Amount by Churn Status")
plt.xlabel("Total Transactions Count")
plt.ylabel("Total Transactions Amount")
plt.legend(title="Churn Status")
plt.show()

# Box Plot: Avg Utilization Ratio for Churned vs. Retained Customers
plt.figure(figsize=(6, 5))
sns.boxplot(data=df_bankchurnersunclean, x="Attrition_Flag", y="Avg_Utilization_Ratio", palette=["lightblue", "salmon"])
plt.title("Average Utilization Ratio by Churn Status")
plt.xlabel("Churn Status")
plt.ylabel("Avg Utilization Ratio")
plt.show()

# Correlation Heatmap: Identifying Key Churn Drivers
plt.figure(figsize=(10, 6))
corr_matrix = df_bankchurnersunclean.select_dtypes(include=["number"]).drop(columns=["CLIENTNUM"]).corr()
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, linewidths=0.5)
plt.title("Correlation Heatmap of Features")
plt.show()

# Line Plot: Total Amount Change vs. Total Count Change by Churn Status
plt.figure(figsize=(8, 5))
sns.lineplot(data=df_bankchurnersunclean, x="Total_Amt_Chng_Q4_Q1", y="Total_Ct_Chng_Q4_Q1", hue="Attrition_Flag", palette=["lightblue", "salmon"])
plt.title("Total Amount Change vs. Total Count Change by Churn Status")
plt.xlabel("Total Amount Change (Q4 to Q1)")
plt.ylabel("Total Count Change (Q4 to Q1)")
plt.legend(title="Churn Status")
plt.show()
```

```python
# Decision Tree Visualization: Identify Key Churn Factors
from sklearn.tree import DecisionTreeClassifier, plot_tree
import numpy as np

# Prepare data for modeling
features = ["Total_Trans_Ct", "Total_Trans_Amt", "Avg_Utilization_Ratio", "Months_Inactive_12_mon", "Credit_Limit"]
X = df_bankchurnersunclean[features]
y = np.where(df_bankchurnersunclean["Attrition_Flag"] == "Attrited Customer", 1, 0)  # Convert churn labels to binary

# Train a simple Decision Tree classifier
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X, y)

# Plot the decision tree
plt.figure(figsize=(12, 6))
plot_tree(clf, feature_names=features, class_names=["Retained", "Churned"], filled=True, fontsize=8)
plt.title("Decision Tree for Predicting Churn")
plt.show()

# Scatter Plot with Clustering: Identifying At-Risk Customer Segments
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df_bankchurnersunclean, x="Credit_Limit", y="Total_Trans_Amt", hue=df_bankchurnersunclean["Attrition_Flag"], alpha=0.7, palette=["lightblue", "salmon"])
plt.title("Credit Limit vs. Transaction Amount by Churn Status")
plt.xlabel("Credit Limit")
plt.ylabel("Total Transaction Amount")
plt.legend(title="Churn Status")
plt.show()

# Cohort Analysis Heatmap: Customer Churn Over Time
cohort_data = df_bankchurnersunclean.groupby(["Months_on_book", "Attrition_Flag"]).size().unstack()
cohort_data = cohort_data.div(cohort_data.sum(axis=1), axis=0)  # Normalize percentages

plt.figure(figsize=(10, 6))
sns.heatmap(cohort_data.T, cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Churn Rate Across Customer Tenure (Months on Book)")
plt.xlabel("Months on Book")
plt.ylabel("Churn Status")
plt.show()
```

```python
# Bar Chart: Churn Rate by Gender (Churn Rate as % of Total Population per Gender)
plt.figure(figsize=(6, 4))

# Calculate the count of churned vs. retained customers for each gender
gender_churn_count = df_bankchurnersunclean.groupby(['Gender', 'Attrition_Flag']).size().unstack()

# Calculate the churn rate as % of total population for each gender
gender_churn_rate = gender_churn_count.div(gender_churn_count.sum(axis=1), axis=0) * 100

# Plot the bar chart for churn rate
gender_churn_rate.plot(kind='bar', stacked=False, color=["lightblue", "salmon"], figsize=(6, 4))
plt.title("Churn Rate by Gender (as % of Total Population per Gender)")
plt.xlabel("Gender")
plt.ylabel("Churn Rate (%)")
plt.legend(title="Attrition Flag")
plt.show()
```
